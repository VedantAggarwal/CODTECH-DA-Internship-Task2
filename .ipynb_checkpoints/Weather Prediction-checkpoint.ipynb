```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df = pd.read_csv('seattle-weather.csv')
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['weather']=df['weather'].map({'drizzle':0,'rain':1,'sun':2,'snow':3,'fog':4})
```

```python
df['temp_avg']=(df['temp_max']+df['temp_min'])/2
```

```python
df
```

```python
weath_model = LinearRegression()
weath_model_df = pd.DataFrame()
```

```python
model_data_X = df[['temp_min']]
model_data_Y = df[['temp_max','temp_avg','precipitation','wind','weather']]
```

```python
model_data_Y
```

```python
weath_model_data_X=model_data_X[['temp_min']]
weath_model_data_Y=model_data_Y[['temp_max']]
```

```python
X_train_weath, X_test_weath, Y_train_weath, Y_test_weath = train_test_split(weath_model_data_X, weath_model_data_Y, test_size=0.2, random_state=42)
```

```python
Y_test_weath
```

```python
weath_model.fit(X_train_weath, Y_train_weath)
```

```python
Y_pred_weath = weath_model.predict(X_test_weath)
```

```python
plt.scatter(X_test_weath,Y_test_weath,color='gray')
plt.plot(X_test_weath,Y_pred_weath,color='red',linewidth=2)
plt.show()
```

```python
print('Mean abolute error is:', metrics.mean_absolute_error(Y_test_weath,Y_pred_weath))
print('Mean squared error is:', metrics.mean_squared_error(Y_test_weath,Y_pred_weath))
print('Root mean squared error is:', np.sqrt(metrics.mean_squared_error(Y_test_weath,Y_pred_weath)))
```

```python
X_test_weath.reset_index(inplace=True)
X_test_weath.drop(['index'],axis=1)
Y_test_weath.reset_index(inplace=True)
Y_test_weath.drop(['index'],axis=1)
Y_pred_weath = pd.DataFrame(Y_pred_weath)
Y_pred_weath = Y_pred_weath.rename(columns={0: 'temp_max'})
```

```python
weath_model_df['temp_min_real'] = X_test_weath['temp_min']
weath_model_df['temp_max_real'] = Y_test_weath['temp_max']
weath_model_df['temp_max_pred'] = Y_pred_weath['temp_max']
weath_model_df
```
